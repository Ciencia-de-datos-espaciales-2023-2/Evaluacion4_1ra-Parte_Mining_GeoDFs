## Ejercicio 1

<div class="alert-success">

2. Compute the distance matrix for both maps.

3. Select one row of the distance matrix, and plot the two points with the minimal distance on top of the country of your choosing.

**1.1. Mapas de puntos a emplear:** 
- Plantas de energía solar en China (China3415_SolarPowerPlants)
- Plantas de energía eólica en China (China3415_WindPowerPlants)

- Ambos se encuentran proyectados y presentan un CRS de 3415.

Fuente empleada: [Global Power Plant Database](https://datasets.wri.org/dataset/globalpowerplantdatabase). Estos datos fueron preprocesados en otro Jupyter Notebook.


```python
import geopandas as gpd
import os

China3415_PowerPlants = "https://github.com/Ciencia-de-datos-espaciales-2023-2/Evaluacion4_1ra-Parte_Mining_GeoDFs/raw/main/maps/China_3415/China3415_PowerPlant.gpkg"

China3415_SolarPowerPlants = gpd.read_file(China3415_PowerPlants,layer="Solar")
China3415_SolarPowerPlants.head()
```

```python
China3415_WindPowerPlants = gpd.read_file(China3415_PowerPlants,layer="Wind")
China3415_WindPowerPlants.head()
```

**1.2. Matriz de distancias entre ambos mapas**

Nota: Ambos mapas presentan proyecciones en metros. Pero, la matriz de distancias se encuentra en unidades de kilómetros y recopila distancias entre una planta de energía eólica a una de energía solar.

```python
China3415_SolarPowerPlants_n = China3415_SolarPowerPlants.set_index("SolarPowerPlant_Name")
China3415_WindPowerPlants_n = China3415_WindPowerPlants.set_index("WindPowerPlant_Name")

DistanceMatrix_WindSolarPlants = China3415_WindPowerPlants_n.geometry.apply\
(lambda g: China3415_SolarPowerPlants_n.geometry.distance(g)/1000)

# Visualización de las 10 primeras filas de la matriz de distancias:
DistanceMatrix_WindSolarPlants.head(10)
```

**1.3. Seleccionar una fila de la matriz de distancias y graficas dos puntos con la mínima distancia sobre el mapa del país.**

Se ha seleccionado la fila correspondiente con la planta de energía eólica "Abag Banner Huiteng Liang Phase 1", se determinó a cuál planta de energía solar se encontraba más cerca y ambos puntos se graficaron en un mapa interactivo de China.

```python
# Planta de energía solar más cercana a la planta eólica "Abag Banner Huiteng Liang Phase 1"
AbagBanner_ClosestSolarPlant = DistanceMatrix_WindSolarPlants.loc["Abag Banner Huiteng Liang Phase 1"].idxmin(axis=0)

#Ambos puntos:
Abag_Banner = China3415_WindPowerPlants[
    China3415_WindPowerPlants.WindPowerPlant_Name=="Abag Banner Huiteng Liang Phase 1"]

AbagBanner_ClosestSolarPlant_p = China3415_SolarPowerPlants[
    China3415_SolarPowerPlants.SolarPowerPlant_Name == AbagBanner_ClosestSolarPlant]

#Mapa interactivo con ambos puntos 
base = Abag_Banner.explore(color="blue",zoom_start =3.5, marker_kwds=dict(radius=3))
AbagBanner_ClosestSolarPlant_p.explore(m=base, color="red",marker_kwds=dict(radius=3))
```

**2.1. Mapas de puntos y líneas a emplear:**
- Mapa de puntos: Plantas de energía nuclear en China (China3415_NuclearPowerPlants).

- Mapa de líneas: Arroyos en China (China3415_Streams).

Fuentes empleadas: [Global Power Plant Database](https://datasets.wri.org/dataset/globalpowerplantdatabase), [Stream Network dataset](https://data.apps.fao.org/catalog/iso/dc2a5121-0b32-482b-bd9b-64f7a414fa0d). 

```python
China3415_NuclearPowerPlants = gpd.read_file(China3415_PowerPlants,layer="Nuclear")
China3415_NuclearPowerPlants.head()
```

```python
China3415_GeoMaps = "https://github.com/Ciencia-de-datos-espaciales-2023-2/Evaluacion4_1ra-Parte_Mining_GeoDFs/raw/main/maps/China_3415/China3415_GeoMap.gpkg"
China3415_Streams = gpd.read_file(China3415_GeoMaps,layer="Streams") 
China3415_Streams.head()
```

**2.2. Matriz de distancias entre ambos mapas**

Nota: Ambos mapas presentan proyecciones en metros. Pero, la matriz de distancias se encuentra en unidades de kilómetros.

```python
China3415_NuclearPowerPlants_n = China3415_NuclearPowerPlants.set_index("NuclearPowerPlant_Name")
China3415_Streams_n = China3415_Streams.set_index("Stream_UniqueArcIdentifier_ARCID")

DistanceMatrix_StreamNuclearPlant = China3415_Streams_n.geometry.apply\
(lambda g: China3415_NuclearPowerPlants_n.geometry.distance(g)/1000)

# Visualización de las 10 primeras filas de la matriz de distancias:
DistanceMatrix_StreamNuclearPlant.head(10)
```

**2.3. Seleccionar una fila de la matriz de distancias y graficar el punto/línea más cercano y lejano.**

Se ha seleccionado la fila correspondiente al arroyo con identificador ARCID 146955, se determinó a cuál planta de energía nuclear se encontraba más cerca y lejos. Los dos puntos y la línea se graficaron en un mapa interactivo de China.

Nota: El arroyo se ve como una línea muy pequeña, porque su masa de agua es menor comparada con la de un río.

```python
# Planta de energía nuclear más cercana y lejana al arroyo con identificador ARCID 146955
Stream146955_ClosestNuclearPlant = DistanceMatrix_StreamNuclearPlant.loc[146955].idxmin(axis=0)
Stream146955_FurthestNuclearPlant = DistanceMatrix_StreamNuclearPlant.loc[146955].idxmax(axis=0)

#Coordenadas de ambos puntos y el arroyo con identificador ARCID 146955
Stream146955_ClosestNuclearPlant_p = China3415_NuclearPowerPlants[
    China3415_NuclearPowerPlants.NuclearPowerPlant_Name==Stream146955_ClosestNuclearPlant]

Stream146955_FurthestNuclearPlant_p = China3415_NuclearPowerPlants[
    China3415_NuclearPowerPlants.NuclearPowerPlant_Name==Stream146955_FurthestNuclearPlant]

Stream146955_p = China3415_Streams[China3415_Streams.Stream_UniqueArcIdentifier_ARCID==146955]

#Mapa interactivo con ambos puntos y el arroyo con identificador ARCID 146955
base = Stream146955_p.explore(color="blue",zoom_start=4)
Stream146955_ClosestNuclearPlant_p.explore(m=base, color="red",marker_kwds=dict(radius=3))
Stream146955_FurthestNuclearPlant_p.explore(m=base, color="green",marker_kwds=dict(radius=3))
```

**3.1. Set de puntos y polígonos a emplear:**
- Set de puntos: Plantas de energía nuclear en China (China3415_NuclearPowerPlants).Es el mismo que se empleó en la pregunta 2.

- Set de polígonos: Convex Hulls de arroyos en China que pertenecen a una misma cuenca principal (China3415_MajorStreamBasin_Hulls). El GeoDataFrame con los convex hulls ya fue creado en el Jupyter Notebook de preprocesamiento.


```python
China3415_MajorStream_BasinsHulls = gpd.read_file(China3415_GeoMaps,layer="MajorStream_BasinsHulls") 
China3415_MajorStream_BasinsHulls.head()

# Notar que "Major_Basin" alude a la cuenca principal donde se ha agrupado un conjunto de arroyos
```

**3.2. Matriz de distancias entre ambos mapas**

Nota: La matriz de distancias se encuentra en unidades de kilómetros.

```python
China3415_MajorStream_BasinsHulls_n = China3415_MajorStream_BasinsHulls.set_index("Major_Basin")

DistanceMatrix_StreamBasinsHulls_NuclearPlant = China3415_MajorStream_BasinsHulls_n.geometry.apply\
(lambda g: China3415_NuclearPowerPlants_n.geometry.distance(g)/1000)

# Visualización de las 10 primeras filas de la matriz de distancias:
DistanceMatrix_StreamBasinsHulls_NuclearPlant.head(10)
```

**3.3. Seleccionar un polígono y graficar el punto más cercano y lejano.**

Se ha seleccionado la fila correspondiente al polígono de la cuenca "Huang He", se determinó a cuál planta de energía nuclear se encontraba más cerca y lejos. Los dos puntos y el polígono se graficaron en un mapa interactivo de China.

```python
# Planta de energía nuclear más cercana y lejana al polígono de la cuenca "Huang He"
HuangHeHull_ClosestNuclearPlant = DistanceMatrix_StreamBasinsHulls_NuclearPlant.loc["Huang He"].idxmin(axis=0)
HuangHeHull_FurthestNuclearPlant = DistanceMatrix_StreamBasinsHulls_NuclearPlant.loc["Huang He"].idxmax(axis=0)

#Coordenadas de ambos puntos y el polígono
HuangHeHull_ClosestNuclearPlant_p = China3415_NuclearPowerPlants[
    China3415_NuclearPowerPlants.NuclearPowerPlant_Name==HuangHeHull_ClosestNuclearPlant]

HuangHeHull_FurthestNuclearPlant_p = China3415_NuclearPowerPlants[
    China3415_NuclearPowerPlants.NuclearPowerPlant_Name==HuangHeHull_FurthestNuclearPlant]

HuangHeHull = China3415_MajorStream_BasinsHulls[China3415_MajorStream_BasinsHulls.Major_Basin=="Huang He"]

#Mapa interactivo con ambos puntos y el polígono
base = HuangHeHull.explore(color="blue",zoom_start=4)
HuangHeHull_ClosestNuclearPlant_p.explore(m=base, color="red",marker_kwds=dict(radius=3))
HuangHeHull_FurthestNuclearPlant_p.explore(m=base, color="green",marker_kwds=dict(radius=3))
```

**4.1. Set de puntos y polígonos a emplear:**
- Set de puntos: Plantas de energía nuclear en China (China3415_NuclearPowerPlants). Es el mismo que se empleó en las preguntas 2 y 3.

- Set de líneas: Conjunto de arroyos en China que pertenecen a una misma cuenca principal (China3415_MajorStreamBasin). El GeoDataFrame con las multilíneas correspondientes a un cuenca principal ya fue creado en el Jupyter Notebook de preprocesamiento.


```python
China3415_MajorStream_Basins = gpd.read_file(China3415_GeoMaps,layer="MajorStream_Basins") 
China3415_MajorStream_Basins.head()

# Notar que "Major_Basin" alude a la cuenca principal donde se ha agrupado un conjunto de arroyos
```

**4.2. *Buffers* alrededor de la cuenca China Coast**

Se crearon *buffers* de los arroyos que forman parte de la cuenca "China Coast". Se seleccionó esta cuenca, porque se localiza al sureste de China, donde la mayoría de las plantas nucleares también se encuentran.

Primero, con la finalidad de evaluar qué rango de distancias emplear para crear los *buffers*, se creó una matriz de distancias:

```python
China3415_MajorStream_Basins_n = China3415_MajorStream_Basins.set_index("Major_Basin")
DistanceMatrix_StreamBasins_NuclearPlant = China3415_MajorStream_Basins_n.geometry.apply\
(lambda g: China3415_NuclearPowerPlants_n.geometry.distance(g)/1000)

import pandas as pd
ChinaCoast_Distances = pd.Series(list(DistanceMatrix_StreamBasins_NuclearPlant.loc["China Coast"]))
ChinaCoast_Distances.describe()
```

Con los resultados obtenidos de esta evaluación preliminar, se observa que la planta nuclear más cercana a la cuenca "Coast China" está a 4.34 km, mientras que la más lejana, a 620.42 km. Con esto en consideración, se crearán buffers de 5 km y 10 km. No fue posible trabajar con mayores distancias, porque el tiempo de computación requerido era elevado.

```python
ChinaCoastBuffer_5km = China3415_MajorStream_Basins[
    China3415_MajorStream_Basins.Major_Basin=="China Coast"
        ].buffer(distance = 5*1000)

ChinaCoastBuffer_5km 
```

**4.3. Mantener los puntos que se encuentran dentro de cada *buffer* y verificar con cuál se obtienen los resultados más interesantes**

```python
NuclearPowerPlants_WithinBuffer_5km = China3415_NuclearPowerPlants.clip(mask=ChinaCoastBuffer_5km)
NuclearPowerPlants_WithinBuffer_5km
```
